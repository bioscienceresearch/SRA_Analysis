```python
import numpy as np
```

```python
BASE_PATH = '/mnt/1TB_0/Data/Assembly/PRJNA605983/'

path='/mnt/1TB_0/Data/Assembly/PRJNA605983/SRR11092060/magic_blast/'
magic_blast_sam_tail='_gsa_bsl_nt_db_magicBLAST.sam'
f_contigs_file_tail='_gsa_bsl_nt_db_magic_blast_asc_contigs.txt'
```

```python
def get_accession(s):
    accession = None
    row= None
    chunks = s.split("\t")
    if len(chunks)>2:
        try:
            int(chunks[2])
        except:
            accession = chunks[2]
            row = s
    return accession, row
```

```python
def write_contigs(sra, magic_blast_sam_tail, f_contigs_file_tail):
    path = BASE_PATH+sra+'/magic_blast/'
    f_contigs= open(path+sra+f_contigs_file_tail, 'w')
    accessions = []
    with open(path+sra+magic_blast_sam_tail) as f:   
        for line in f:
            if line.startswith('k141'):
                accession, row=get_accession(line)
                if row is not None:
                    accessions.append(accession)
                    f_contigs.write(f'{row}\n')
    return accessions
```

```python
def get_accession_dat(values):
    vdats=[]
    for v in values:
        vdat=!blastdbcmd -db gsa_bsl_nt_db -entry $v
        vdats.append(vdat[0].split(',')[0])
    return vdats
```

```python
def workflow(sra):
    accessions = write_contigs(sra, magic_blast_sam_tail, f_contigs_file_tail)
    values, counts = np.unique(accessions, return_counts=True)
    vdats=get_accession_dat(values)
    return vdats, counts
```
